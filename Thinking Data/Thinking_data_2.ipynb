#### Now from this checking of data we know that in subscription some are lowercase while some upper case 
#### In watch time values are missing and are of mixed type
#### There is duplicated data in movies and names

```python
from word2number import w2n

def cleaned_data(data):

    unique_records = set()
    cleaned_users = []

    for user in data:

        # Subscription Consistency
        user["subscription"] = (
            user["subscription"]
            .strip()
            .lower()
            .capitalize()
        )

        # Watch Time Consistency
        watch_time = str(user["watch_time"]).strip().lower()

        try:
            user["watch_time"] = int(watch_time)

        except:
            try:
                user["watch_time"] = w2n.word_to_num(watch_time)

            except:
                user["watch_time"] = 0

        # Deduplication
        record = (user["user"], user["movie"])

        if record not in unique_records:
            unique_records.add(record)
            cleaned_users.append(user)

    return cleaned_users                   
                    
```

```python
cleaned = cleaned_data(data)

for user in cleaned:
    print(user)
```

```python
def get_insights(data):

    tot_watch_time = 0

    for user in data:
        tot_watch_time += user["watch_time"]

    avg_watch_time = tot_watch_time / len(data)

    print(f"Total Watch Time: {tot_watch_time}")
    print(f"Average Watch Time: {avg_watch_time:.2f}")

    # Premium User Count
    premium_user_count = 0

    for user in data:
        if user["subscription"] == "Premium":
            premium_user_count += 1

    print(f"Premium Users: {premium_user_count}")

    # Most Active User
    max_watch = 0
    active_user = ""

    for user in data:
        if user["watch_time"] > max_watch:
            max_watch = user["watch_time"]
            active_user = user["user"]

    print(f"Most Active User: {active_user}")
    print(f"Watch Time: {max_watch}")

    #Buisiness decision    
    if avg_watch_time >= 120:
        print("Users are highly engaged")
    elif avg_watch_time >= 60:
        print("Moderately engaged")
    else:
        print("Poor engagement")
```

```python
insights=get_insights(data)
```
