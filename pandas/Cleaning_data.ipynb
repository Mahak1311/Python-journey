# Cleaning Data

****
### Handling Missing Values

```python
df= pd.read_csv("raw_data.csv")

df.isnull() #gives true where there is null value
df.isna() #same as isnull()
df.isnull().sum() #to count how many null values are there
df.dropna() #to drop all rows with null values
df.dropna(axis=1) #to drop all column with null values
df.fillna(0) #to fill null values with smth
df["age"]=df["age"].fillna(df["age"].mean())

df.ffill() #forward fill
df.bfill() #backword fill
```

### Handle duplicate values

```python
df = pd.read_csv("raw_data.csv") 

df.duplicated()                      # True for duplicate rows
df.duplicated("country")             # True for duplicate rows in particular col
df.duplicated(["country", "gender"]) 

df.drop_duplicates()                 # drops duplicate rows
```

### Handle data types

```python
df = pd.read_csv("raw_data.csv") 

df.dtypes                                     # dtype of each col
df2 = df.fillna(0).copy()
df2["income"] = df2["income"].astype(int)       # change dtype

#String types
df2["country"].str.lower() #lowers case
df2["country"].str.upper() #uppers case
df2["country"].str.capitalize() #capitalizes words
df2["country"].str.strip() #strips space
df2["name"].str.split(" ") #splits words
df2["country"].str.contains("US") #gives true or false based on if smth exists
df2["country"].str.contains("india", case=False) #in case upper or lower doesn't matter
```

### Transformation/Feature Engineering

```python
df2=df2.copy() #to make a copy
#applying a function on a row or column 
df2["Tax"]=df2["income"].apply(lambda x: "10%" if x <6000 else "20%") #lambda makes function in a single line if x<6000 then tax is 10% otherwise 20%
df2
#map values to other
new_gender_values = {
    "Male" : "M",
    "Female": "F",
    "Unknown" : "U"}
df2["gender"]=df2["gender"].map(new_gender_values)
df2
#assign some new values
df2.assign(new_income=df2["income"] * 1.1) #1.1 means increase by 10%

#replace specific values
df2["country"].replace("USA","United States")
```

### Transformation- Sorting

```python
df2=df.copy()

#Renaming

df2.columns= ["Id","Name","Age","Country","Gender","Income","Tax"]
df2.rename(columns={"Income":"Salary"})
df2.rename(index={0:"first"})

#Sorting-Values & index
df2.sort_values("Income") #in ascending
df2.sort_values("Income", ascending=False) #in descending
df2.sort_values(["Age","Income"]) #sorts on basis of age and if age is same then sorts on basis of income
sorted_df2 =df2.sort_values(["Age","Income"]) #sorts values but not index
sorted_df2.sort_index() #sorts index but values aren't sorted

#Reset Index- sorts both index and value
sorted_df2.reset_index()
sorted_df2.reset_index(drop=True) #to drop original value

#Ranking
df2["Ranking"] = df2["Income"].rank(ascending=False,method="dense") #by dense method same income values get same rank 
#while in default method it gives average rank and in min method it skips next rank 

#Reorder
df2[["Id", "Name", "Age", "Income","Tax", "Gender", "Country", "Ranking"]]
```

### Writing to csv files

```python
df = pd.read_csv("raw_data.csv")

df2= df.copy()

df2.drop_duplicates(inplace=True)  #inplace means to change in original df
df2.sort_values("income",inplace=True)

df.to_csv("sorted_data.csv") #save df as a csv file
df.to_csv("sorted_data.csv", index=False) #removes index
json_string = df.to_json() #converts to json
print(json_string)
```

### Grouping & Aggregation

```python
df2=df.copy()

df2.groupby("country")["income"].mean() #gives mean of income acc to country

df2.groupby("gender")["income"].mean()  #mean of income acc to gender

df2.groupby("gender")["income"].max()  #max of income acc to gender

df2.groupby("gender")["income"].min()  #min of income acc to gender

df2.groupby("country")["gender"].count()   #count of gender acc to country

df2.groupby("country")["income"].agg(["mean","max","min"]) #uses aggregation to apply more than 1 property

df2.groupby("country")["income"].agg(avg_salary="mean",max_salary="max",min_salary="min") #gives name 

df2.groupby("country").agg({ #aggregate on multiple cols
    "age":"mean",
    "income":"max"
})
df2.groupby("country").agg(
    avg_age=("age","mean"),
    max_income=("income","max") #give them name
)
```
