```python
data=pd.read_csv("globalAirQuality.csv")
print(data)
```

```python
data
```

```python
#selecting data

#selecting columns
data["country"]
data[["city","aqi"]]

#selecting row-by label
data.loc[0]
data.loc[0:4] #4 is inclusive in label

#selecting row-by index
data.iloc[0]
data.iloc[0:7] #7 is not inclusive

#select row and column
data.loc[0,"country"] #can contain string
data.iloc[0,2] #only numbers

#slicing
data.loc[0, ["country","aqi"]]
data.iloc[0,1:5]


data.loc[0:5, ["country", "city", "latitude"]]   # end is inclusive
data.iloc[0:5, 1:3]                              # end is exclusive 

# Select single scalar value
data.at[0, "city"]
data.iat[0, 2]
```

```python
#filtering
data[data["aqi"]>100]
data[data["aqi"]>100][["city","aqi"]] #here both rows and columns are filtered

data[(data["aqi"]>100)&(data["timestamp"] =="2025-11-04 18:25:17.554219")] #here only rows are filtered

aqi_data=data[(data["aqi"]>100)&(data["timestamp"] =="2025-11-04 18:25:17.554219")][["city","aqi"]]


#Query- returns a copy not a view
data.query("aqi > 100 & timestamp == '2025-11-04 18:25:17.554219'")

my_country = "IN"
data.query("country == @my_country") #@ is used to refernce 

data["timestamp"]=pd.to_datetime(data["timestamp"])
data["date"] = data["timestamp"].dt.date #pandas read them as objects(strings)
selected_date = pd.to_datetime("2025-11-04").date() #convert them so now pandas cann recognise it

data.query("country == @my_country & date == @selected_date & aqi > 100")

#to create copy
countries=data[["country","aqi"]].copy()
countries.loc[0:4, "country"] = "India"

print(countries)
```
